**Let's dive into the age data**

```python
data.age.unique()
```

```python
#Often the numerical value has a "s" at the end to mark it's year range. We will remove it to keep a numerical type. 

age_v1=[]
for e in data.age:
    age_v1.append(str(e.replace('s','').lower()))
```

```python
#Through regex I will put as unknown all none numerical values which does not contains "teen"

import re

age_v2 = []
for e in age_v1:
    if 'teen' in e:
        age_v2.append('teen')
        

    elif re.search(r'^[0-9]+$', str(e)):
        age_v2.append(e)
    else:
        age_v2.append('unknown')
```

```python
age_v3 = []
for e in age_v2:
    if e == 'unknown':
        age_v3.append('unknown')
    elif e == 'teen':
        age_v3.append('10s')
    elif int(e) < 10:
        age_v3.append('0s')
    elif int(e) <20:
        age_v3.append('10s')
    elif int(e) <30:
        age_v3.append('20s')
    elif int(e) <40:
        age_v3.append('30s')
    elif int(e) <50:
        age_v3.append('40s')
    elif int(e) <60:
        age_v3.append('50s')
    elif int(e) <70:
        age_v3.append('60s')
    elif int(e) <80:
        age_v3.append('70s')
    elif int(e) <90:
        age_v3.append('80s')
    else: 
        age_v3.append('90s')        
```

```python
data['Range_Age'] = age_v3
data.age.unique()
```

**Cleaning Time column**

```python
data.time.unique()
```

```python
time_v2 = []
for e in data.time:
    time_v2.append(str(e).replace('h',':'))
```

```python
import re

# Initialize empty list to store cleaned data
time_v3 = []

# Iterate through each element in the data list
for e in time_v2:    
    match = re.search(r"^([0-9]{2}):([0-9]{2})$", e)
    if match:
        time_v3.append(e)
    else:
        time_v3.append("Unknown")
```

```python
time_v4=[]
for e in time_v3:
    if e == 'Unknown':
        time_v4.append('Unknown')
    else :
        time_v4.append(str(e).replace(':','h')[:3])
```

```python
data['Time_Range'] = time_v4
```
